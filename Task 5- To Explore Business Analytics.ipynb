```python
df["Category"].unique()
```

```python
s_group = df.groupby('Category')['Sales'].sum().reset_index()
print(s_group)
```

```python
labels = df['Category'].unique()
plt.figure(figsize = (7, 7))
colors = ['deepskyblue', 'steelblue', 'cadetblue']

plt.pie(s_group['Sales'], autopct='%1.1f%%',labels = labels, explode = (0.03,0.03,0.05), colors = colors)
plt.title('Sales with respect to Category', fontweight = 'bold', size = 15)

plt.show()
```

```python
profit = df[df['Profit'] > 0]
neutral = df[df['Profit'] == 0]
loss = df[df['Profit'] < 0]
```

```python
p_group = profit.groupby('Category')['Profit'].sum().reset_index()
print(p_group)
```

```python
labels = df['Category'].unique()
plt.figure(figsize = (6, 6))
colors = ['violet', 'purple', 'mediumvioletred']

plt.pie(p_group['Profit'],autopct='%1.1f%%',labels = labels, explode = (0.03,0.03,0.05), colors = colors)
plt.title('Profit with respect to Category', fontweight = 'bold', size = 15)

plt.show()
```

```python
rp_group = profit.groupby('Region')['Profit'].sum().reset_index()
print(rp_group)
```

```python
labels = df['Region'].unique()
plt.figure(figsize = (6, 6))
#colors = ['red', 'orangered', 'tomato', 'darkred']

plt.pie(rp_group['Profit'],autopct='%1.1f%%',labels = labels, explode = (0.03,0.03,0.03, 0.03))
plt.title('Profit with respect to Region', fontweight = 'bold', size = 15)
plt.show()
```

```python
sp_group = profit.groupby('Segment')['Profit'].sum().reset_index()
print(sp_group)
```

```python
labels = df['Segment'].unique()
plt.figure(figsize = (6, 6))
#colors = ['forestgreen', 'seagreen', 'springgreen']

plt.pie(sp_group['Profit'],autopct='%1.1f%%',labels = labels, explode = (0.03,0.03, 0.03))
plt.title('Profit with respect to Segment', fontweight = 'bold', size = 15)
plt.show()
```

```python
profit.head()
```

```python
disc = profit[profit['Discount'] > 0]
w_disc = profit[profit['Discount'] == 0]
```

```python
labels = ['Discount', 'Without Discount']
size = [disc['Profit'].mean(), w_disc['Profit'].mean()]
plt.figure(figsize = (6, 6))

plt.pie(size, autopct='%1.1f%%', labels = labels, explode = (0.03, 0.03))
plt.title('Profit with respect to Discount', fontweight = 'bold', size = 15)
plt.show()
```

```python
sscb_group = profit.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by = 'Profit', ascending=True)
```

```python
sscb_group
```

```python
plt.figure(figsize = (20, 8))

pcat = sns.barplot('Sub-Category', 'Profit', data = sscb_group)

for p in pcat.patches:
    pcat.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
plt.title('Profit with respect to Sub-Category', fontweight = 'bold', size = 15)
plt.show()

```

# Profit:

We got highest sales and profit from Technology. Interestingly, we got highest profit by selling less number of items from Technology.

As our customers are mostly from the West side, we got most of our profit from there. Our profit got decreased, when we offer any discount on a product. We have got 64.6% of our profit from the products without discount and only 36.4% of our profit from the products with discount
