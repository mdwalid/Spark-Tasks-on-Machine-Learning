```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('full_data.csv',parse_dates= ['date'], index_col = 'date')
```

```python
df.tail()
```

```python
df.shape
```

```python
wrc = df.groupby('location')['new_cases'].sum().reset_index().sort_values(by = 'new_cases', ascending=False)
print(wrc)

```

# Top 10 covid affected countries in the world

```python
top_10=wrc[1:10] 
```

```python
top_10
```

```python
labels = top_10['location'].unique()
plt.figure(figsize = (8, 8))

plt.pie(top_10['new_cases'], autopct = '%1.1f%%', labels = labels)

plt.title('Top 10 covid affected countries in the world', size = 15, fontweight = 'bold')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize = (15, 8))

pcat = sns.barplot('location', 'new_cases', data = top_10)

for p in pcat.patches:
    pcat.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
plt.title('Top 10 covid affected countries in the world', fontweight = 'bold', size = 15)
plt.show()

```

# Top 10 countries in the world with respect to new_deaths

```python
wrc = df.groupby('location')['new_deaths'].sum().reset_index().sort_values(by = 'new_deaths', ascending=False)
print(wrc)
```

```python
top_10_d= wrc[1:11]
top_10_d
```

```python
import seaborn as sns

plt.figure(figsize = (15, 8))

pcat = sns.barplot('location', 'new_deaths', data = top_10_d)

for p in pcat.patches:
    pcat.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
plt.title('countries with respect to new_deaths', fontweight = 'bold', size = 15)
plt.show()
```
